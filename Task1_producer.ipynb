```python
##Q1

# import statements
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt
from datetime import datetime,timezone,timedelta
from pytz import timezone
import csv

def read_csv(fileName):
    '''Read the CSV file clickstream.csv'''
    '''you can use csv.DictReader'''
    list = []
    with open(fileName, 'rt') as f:
        reader = csv.DictReader(f)
        for row in reader:
            list.append(row)
    return list

def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, data)
        print('Message published successfully. Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
    
    
if __name__ == '__main__':
    
    #Topic names
    topic1 = 'Bureau_data'
    topic2 = 'Customer_Data'
    
    #Reading the data
    cRows = read_csv('bureau.csv')
    cRows1 = read_csv('customer.csv')
    
    print('Publishing records..')
    producer = connect_kafka_producer()
    
    #records counter
    t=0
    k = True
    while(k!=False):
        
        #Generating random numbers 
        r = random.randint(10,30)
        
        #Creating timestamp
        ts = int(datetime.now(timezone('UTC')).timestamp())
        
        if t+r > len(cRows1):
            t=0
            
        #taking the required rows
        a = cRows1[t:t+r]
        
        #Updating counter
        t=t+r
        
        #adding timestamp for each row
        for i in a:
            i['ts'] = ts
            
        #Publish customer data
        publish_message(producer, topic2, a)
        
        b=[]
        
        #Getting the required bureau rows and the same timestamp
        for j in a:
            b = b + ([d for d in cRows if d['ID'] == j['ID']])
        for j in range(1,len(b)):
            b[j]['ts'] = ts
        
        #Publishing the bureau data
        publish_message(producer, topic1, b)
        sleep(5)
```
